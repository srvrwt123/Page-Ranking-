```python
def qid_ex(qid):
    return qid[4:] # returns the value after the query id i.e. the value.

def val_ex(ve):
    return ve.split(':')[1] # splits the queries with the ':'.
```

```python
val[1] = val[1].apply(qid_ex)
# applies map to each element in the qid_ex i.e. after 4th index.
val[val.columns[2:-1]] = val[val.columns[2:-1]].applymap(val_ex)
#it shows only the values of the features and not the feature id and is applied to each and every element.  
val.drop( 138, axis = 1)
#remove the last column as it only contains NaN values.
```

```python
def df_transf(df):
    df[1] = df[1].apply(qid_ex)
    df[df.columns[2:-1]] = df[df.columns[2:-1]].applymap(val_ex)
    df = df.drop(138, axis=1)
    return df
```

```python
#applies the above function to the dataset which is the file train.txt.
training_data_frame = df_transf(train) 
```

Let's perform some optimization techniques 

```python
# Discounted Cumulative Gain
import math
def dcg(data):
    n = len(data)
    while sum(data)==0:
        return 0
    else:
        value_1 = range(0, n)
        t = 0
        for i in value_1:
            clt_calc=(2**(data[i])-1)
            cr_t= clt_calc*(math.log((float(i)+2), 2)**(-1))
            t = t + cr_t
        return(t)
```

```python
dcg(prediction_real_val) # DCG prediction value
```

```python
def dcg1(od_grp):
    od_grp = od_grp.sort_values(by = "predicted").real_val.tolist() # sorts it with the real value and give output with the query id.
    return dcg(od_grp)
```

```python
dcg_score = gt_val.apply(dcg1)
```

```python
#Normalized Discounted Cululative Gain
def ndcg(data, n):
    if sum(data)==0:
        return 0
    else:
        val_2 = range(0, n)
        DCG = 0
        for i in val_2:
            clt_calc=(2**(data[i])-1)
            clt_calc1= clt_calc*(math.log((float(i)+2), 2)**(-1))
            DCG = DCG + clt_calc1
        data_sort= sorted(data,reverse=True)
        y = len(data)
        val_2 = range(0, y)
        iDCG = 0
        for i in val_2:
            clt_calc=(2**(data_sort[i])-1)
            clt_calc1= clt_calc*((math.log((i+2), 2))**(-1))
            iDCG = iDCG + clt_calc1
        return(DCG/iDCG)
```
